```python
#kaggleデータの確認
import pandas as pd

# CSV の読み込み
df = pd.read_csv("../../data/raw/kaggle/19860105-20210731_race_result.csv")

# レース日付の列名が "レース日付" または "date" の可能性があるので対応
date_col_candidates = ["レース日付", "race_date", "date"]

# 存在する日付カラムを特定
for col in date_col_candidates:
    if col in df.columns:
        date_col = col
        break
else:
    raise ValueError("日付列が見つかりません。列名を確認してください。")

# 日付型に変換
df[date_col] = pd.to_datetime(df[date_col])

# 最新順にソートして上位10件
latest_10 = df.sort_values(date_col, ascending=False).head(10)

latest_10
```

```python
import pandas as pd

# CSV読み込み
df = pd.read_csv("../../data/raw/kaggle/19860105-20210731_race_result.csv")

# レースIDの列名候補
race_id_candidates = ["レースID", "race_id", "raceId", "id"]

# 実際に存在する列名を特定
for col in race_id_candidates:
    if col in df.columns:
        race_id_col = col
        break
else:
    raise ValueError("レースID列が見つかりません。列名を確認してください。")

# レースIDで抽出
target_race_id = "202104030302"
filtered = df[df[race_id_col] == target_race_id]

filtered
```

```python
import pandas as pd

df = pd.read_csv("../../data/processed/kaggle/lgbm_win_pred.csv")

# ① 単勝オッズのレンジ確認（倍率として 1〜100 くらいに収まるはず）
print(df["単勝"].describe())

# ② 全頭均等BETしたときのROI
total_bet = 100 * len(df)
total_return = (df["target_win"] * df["単勝"] * 100).sum()
print("naive_roi =", total_return / total_bet)
```
